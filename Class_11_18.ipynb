```python
def histogram(values, bin_width):
    if len(values) == 0:
        return []
    
    large = max(values)
    #largest number total of all bins can hold is 
    #number of bins * bin width -1
    # 5 bins bin width 5 24
    # 4 bins bin width 2 7
    bincount = large//bin_width+1
    bincount = dict.fromkeys(range(bincount), 0) #keys will be bin num and value will be count
    for val in values:
        bin = val//bin_width
        bincount[bin] += 1
    return list(bincount.values())
```

```python
histogram([1, 1, 0, 1, 3, 2, 6], 1)
```
